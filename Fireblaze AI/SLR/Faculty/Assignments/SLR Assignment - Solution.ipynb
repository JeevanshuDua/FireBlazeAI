<span style='color:Blue'> **Question - 5.How to declare the characteristic variable as well as the target variable.** </span>

```python
# Declare feature variable and target variable

X = dataset['Sales'].values
y = dataset['Advertising'].values

# Sales and Advertising data values are given by X and y respectively.

# Values attribute of pandas dataframe returns the numpy arrays.
```

<span style='color:Blue'> **Question - 6. How dependent and independent variables relate to each other?** </span>

<span style='color:Blue'> **Question - 7. Check sales against advertising.** </span>

NOTE:- Plot the bar plot.

```python
# import seaborn 
import seaborn as sns


# set the figure size
plt.figure(figsize = (15,8))

# plot a barplot with mean age on y-axis and income on x-axis
sns.barplot(x = 'Sales', y = 'Advertising', data = dataset)

# add plot and axes labels
# set text size using 'fontsize'
plt.title('Bar Plot for Sales and Advertising', fontsize = 15)
plt.xlabel('Sales', fontsize = 15)
plt.ylabel('Advertising', fontsize = 15)

# display the plot
plt.show()
```

<span style='color:Blue'> **Question - 8. Seperate out the dependent and independent variables.** </span>



```python
X = dataset.drop(["Sales"],axis=1)
y = dataset.Sales
```

<span style='color:Blue'> **Question - 9. Divide the data in proportion 33%** </span>

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
# Split X and y into training and test data sets

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.33, random_state=42)

```

```python
# Print the dimensions of X_train,X_test,y_train,y_test

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

<span style='color:Blue'> **Question - 10. Load and fit the model according to the dependant variable.** </span>

```python
# Fit the linear model

# Instantiate the linear regression object lm
from sklearn.linear_model import LinearRegression
lm = LinearRegression()


# Train the model using training data sets
lm.fit(X_train,y_train)
```

<span style='color:Blue'> **Question - 11. Predict the outcome of the test.** </span>

```python
# Predict on the test data
y_pred=lm.predict(X_test)
y_pred
```

<span style='color:Blue'> **Question - 12. What is the regression coefficient for given model?** </span>

```python
# regression coefficients
print('Coefficients: ', lm.coef_)
```

<span style='color:Blue'> **Question - 13. How to compute model incline and intercept?** </span>

```python
# Compute model slope and intercept

a = lm.coef_
b = lm.intercept_,
print("Estimated model slope, a:" , a)
print("Estimated model intercept, b:" , b) 
```


<span style='color:Blue'> **Question - 14. Predicting the Advertising values on first five Sales datasets?** </span>

```python
# Predicting Advertising values
lm.predict(X)[0:5]

# Predicting Advertising values on first five Sales values.
```


<span style='color:Blue'> **Question - 15. How can we tell that this model is perfectly predicted?** </span>

```python
#variance score: 1 means perfect prediction
print('Variance score: {}'.format(lm.score(X_test, y_test)))
```


<span style='color:Blue'> **Question - 16. How indentify our model it good or not? Which method used?** </span>

NOTE:- Find out error

```python
# Calculate and print Root Mean Square Error(RMSE)

from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
print("RMSE value: {:.4f}".format(rmse))
```

```python
# Calculate and print r2_score

from sklearn.metrics import r2_score
print ("R2 Score value: {:.4f}".format(r2_score(y_test, y_pred)))
```

## <span style='color:Red'> **2. Mean of Residuals** </span>

* Residuals as we know are the differences between the true value and the predicted value.

<span style='color:Blue'> **Question - 20. Find out difference between the true value and predicted value.** </span>

```python
residuals = y_test - y_pred
mean_residuals = np.mean(residuals)
print("Mean of Residuals {}".format(mean_residuals))
```

* Very close to zero so all good here.

## <span style='color:Red'> **3.  Check for Homoscedasticity** </span>

* Homoscedasticity means that the residuals have equal or almost equal variance across the regression line.

### Detecting heteroscedasticity!

* Graphical Method: Firstly do the regression analysis and then plot the error terms against the predicted values( Yi^). If there is a definite pattern (like linear or quadratic or funnel shaped) obtained from the scatter plot then heteroscedasticity is present.

<span style='color:Blue'> **Question - 21. Show the residuals have equal variance across the regression line**</span>

NOTE:- Plot the homoscedasticity

```python
p = sns.scatterplot(y_pred, residuals, data=dataset)
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot([0,26],[0,0],color='blue')
p = plt.title('Residuals vs fitted values plot for homoscedasticity check')
```

## <span style='color:Red'> **4.  Check for Normality of error terms/residuals** </span>

<span style='color:Blue'> **Question - 22. Show the normality of residuals terms.**</span>

```python
p = sns.distplot(residuals,kde=True)
p = plt.title('Normality of error terms/residuals')
```

## <span style='color:Red'> **5. No autocorrelation of residuals** </span>

* When the residuals are autocorrelated, it means that the current value is dependent of the previous (historic) values and that there is a definite unexplained pattern in the Y variable that shows up in the error terms. 

<span style='color:Blue'> **Question - 23. Is there any autocorrelation? show by visualization**</span>

```python
plt.figure(figsize=(10,5))
p = sns.lineplot(y_pred,residuals,marker='o',color='blue')
plt.xlabel('y_pred/predicted values')
plt.ylabel('Residuals')
plt.ylim(-10,10)
plt.xlim(0,26)
p = sns.lineplot([0,26],[0,0],color='red')
p = plt.title('Residuals vs fitted values plot for autocorrelation check')
```
